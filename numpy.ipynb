```python
#Giải hệ
# x + 2y + 3z= 14
# 2x+5y+z = 18
# 4x+6y+8z=32

#1
A = np.array([[1,2,3],[2,5,1],[4,6,8]], dtype=float)
b= np.array([14,18,32],dtype=float)
x= np.linalg.solve(A,b)
print(x)
print(A @ x)
#2
M = np.array([[4, -2], [1, 1]])
eigenvalues, eigenvectors = np.linalg.eig(M)

for i in range(len(eigenvalues)):
    lam = eigenvalues[i].real
    v = eigenvectors[:, i].real
    print(M @ v, "~", lam * v)
```
